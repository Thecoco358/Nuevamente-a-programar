```python
import random

opciones = ["piedra", "papel", "tijeras"]

print(" JUEGO: PIEDRA, PAPEL O TIJERAS ")

usuario = input("Elige una opción (piedra, papel, tijeras): ").lower()

if usuario not in opciones:
    print("Opción inválida. Intenta de nuevo.")
else:
    computadora = random.choice(opciones)
    print(f"La computadora eligió: {computadora}")

    if usuario == computadora:
        print("¡Empate!")
    elif (usuario == "piedra" and computadora == "tijeras") \
        or (usuario == "papel" and computadora == "piedra") \
        or (usuario == "tijeras" and computadora == "papel"):
        print("¡Ganaste!")
    else:
        print("Perdiste :(")
```
